```python
import pandas as pd
import matplotlib as plt
import seaborn as sns

motor_ab = pd.read_csv('abnormal.csv')
motor_no = pd.read_csv('normal.csv')                       
```

```python
motor_ab
```

```python
motor_train_ab = motor_ab[motor_ab['motor_id'] <= 36]
motor_test_ab = motor_ab[motor_ab['motor_id'] > 36]
motor_train_no = motor_no[motor_no['motor_id'] <= 36]
motor_test_no = motor_no[motor_no['motor_id'] > 36]
```

```python
motor_train = pd.concat([motor_train_ab, motor_train_no], ignore_index=True)
motor_test = pd.concat([motor_test_ab, motor_test_no], ignore_index=True)
```

```python
motor_train = motor_train.drop("motor_id", axis=1)
```

```python
motor_test = motor_test.drop("motor_id", axis=1)
```

```python
motor_train.to_csv('motor_train.csv', index=False, encoding='utf-8-sig')
motor_test.to_csv('motor_test.csv', index=False, encoding='utf-8-sig')
```
